```python
location_data = pd.DataFrame(columns= ["user_id", "location"])
```

```python
iteration_size = 100
index = 0
start = 0
end = 0
while end < len(user_id):
    index += 1
    start = end
    if start + iteration_size > len(user_id):
        end = len(user_id)
    else:
        end = start + iteration_size
      
    print (index, start, end, (end-start) )
    
    users_data = api.lookup_users(user_ids = list(user_id[start:end]))
    
    users_data = [[t.id, t.location] for t in users_data]
    
    if index == 1:
        location_data = pd.DataFrame(users_data, columns=["user_id", "location"])
        
    else:
        df = pd.DataFrame(users_data, columns=["user_id", "location"])
        location_data = location_data.append(df)
```
