```python
# Handling Emoji patterns:

emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"
         "]+", flags=re.UNICODE)

# Set of Happy Emoticons:

emoticons_happy = set([':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'])

# Set of Sad Emoticons:

emoticons_sad = set([':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('])

emoticons = emoticons_happy.union(emoticons_sad)
```

```python
def remove_url(tweet):
    return re.sub(r"http\S+", "", tweet)

def clean_tweets(tweet):
    
    # First Pass of Pre-processing:
 
    tweet = remove_url(tweet)
    
    stop_words = set(stopwords.words('english'))
    word_tokens = nltk.word_tokenize(tweet)

    #after tweepy preprocessing the colon symbol left remain after removing mentions:
    
    tweet = re.sub(r':', '', tweet)
    tweet = re.sub(r'‚Ä¶', '', tweet)

    # Replacing consecutive non-ASCII characters with a space:
    
    #tweet = re.sub(r'[^\x00-\x7F]+',' ', tweet)
    tweet = tweet.strip()
    
    # Replacing 2+ dots with space
    
    tweet = re.sub(r'\.{2,}', ' ', tweet)
    
    # Stripping space, " and ' from tweet
    
    tweet = tweet.strip(' "\'')

    # Removing emojis from tweet using earlier emoji lists:
    
    tweet = emoji_pattern.sub(r'', tweet)
    
    # Replacing multiple spaces with a single space:
    
    tweet = re.sub(r'\s+', ' ', tweet)
    
    # Filtering using string appending:
    
    filtered_tweet = [w for w in word_tokens if not w in stop_words]
    filtered_tweet = []
       
    for w in word_tokens:
        if w not in stop_words and w not in emoticons and w not in string.punctuation:
            filtered_tweet.append(w)
            
    return ' '.join(filtered_tweet)
```

```python
def clean_tweets1(clean_text):
    
    # Second Pass of Pre-processing:
    
    for i in range(len(clean_text)):
        clean_text[i] = clean_text[i].replace('...', ' ')
        clean_text[i] = clean_text[i].replace('_', '')
        clean_text[i] = clean_text[i].replace(" ’ ", ' ')
        clean_text[i] = clean_text[i].replace(" '", ' ')
        clean_text[i] = clean_text[i].replace("' ", ' ')
        clean_text[i] = clean_text[i].replace("'", ' ')
        clean_text[i] = clean_text[i].replace(" ” ", ' ')
        clean_text[i] = clean_text[i].replace(" “ ", ' ')
        clean_text[i] = clean_text[i].replace("“ ", ' ')
        clean_text[i] = clean_text[i].replace(" “", ' ')
        clean_text[i] = clean_text[i].replace(" ”", ' ')
        clean_text[i] = clean_text[i].replace(". ", ' ')
        clean_text[i] = clean_text[i].replace("/", ' ')
        clean_text[i] = clean_text[i].replace(" `` ", ' ')
        clean_text[i] = clean_text[i].replace(" -- ", ' ')
        clean_text[i] = clean_text[i].replace(" â€˜ ", ' ')
        clean_text[i] = clean_text[i].replace(" â€˜", ' ')
        clean_text[i] = clean_text[i].replace("â€˜ ", ' ')
        clean_text[i] = clean_text[i].replace(".", '')
        clean_text[i] = clean_text[i].replace(". ", ' ')
        clean_text[i] = clean_text[i].replace("â€", '')
        clean_text[i] = clean_text[i].replace(" â€” ", ' ')
        clean_text[i] = clean_text[i].replace("â€” ", ' ')
        clean_text[i] = clean_text[i].replace(" â€”", ' ')
        clean_text[i] = clean_text[i].replace("â€™", '')
        clean_text[i] = clean_text[i].replace(" â€™ ", ' ')
        clean_text[i] = clean_text[i].replace("â€™ ", ' ')
        clean_text[i] = clean_text[i].replace(" â€™", ' ')
        clean_text[i] = clean_text[i].strip()
        clean_text[i] = re.sub(r'[^\x00-\x7F]+',' ', clean_text[i])

        
    return clean_text
```
